```python
def print_individual_col(df, col):
    d = df[col]
    print("Column: '%s'" % col, end='\n\n')
    print('data = %s' % str(list(d.unique())))
    print('length = %d' % len(d.unique()), end='\n\n')
    print(d.value_counts(), end='\n\n\n')
```

```python
for col in eda_cols:
    print_individual_col(df, col)
    print('============================================================')
```

```python
stations = list(df.station_code.unique())
stations.sort()
```

```python
d = {}
month_days = df.day.max()
cols_1h = ['H01','H02', 'H03', 'H04','H05','H06','H07','H08', 'H09','H10','H11','H12',
            'H13','H14', 'H15', 'H16', 'H17','H18', 'H19', 'H20', 'H21','H22','H23', 'H24']

for s in stations:
    s_str = str(s)
    d[s_str] = {}
    d_s = df[df.station_code == s]
    
    c = list(d_s.contaminant_code.unique())
    c.sort()
    
    for ct in c:
        ct_str = str(ct)
        d_c =  d_s[d_s.contaminant_code == ct]
        d[s_str][ct_str] = list()
        
        for day in range(1, month_days + 1):
            d[s_str][ct_str].extend(d_c[d_c.day == day][cols_1h].values.tolist()[0])
    
```
