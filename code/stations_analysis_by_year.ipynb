## Import data

```python
df_2018 = pd.read_csv(stn.STATIONS_BASE_PATH + '/2018.csv')

eda_cols, df_2019 = stn.load_station_data(
    stn.STATIONS_BASE_PATH + '/2019.csv', 
    stn.station_cols)

_, df_2021 = stn.load_station_data(
    stn.STATIONS_BASE_PATH + '/2021.csv', 
    stn.station_cols)

_, df_2022 = stn.load_station_data(
    stn.STATIONS_BASE_PATH + '/2022.csv', 
    stn.station_cols)
```

## Stations in the year 2018

```python
# NO2 seems to be the most important contaminant
# followed by O3 and PM10.

# class_1 = Urban or Suburban (city location)
# class_2 = Emission sources 

cols_2018 = ['station_name', 'code_dtes', 'zqa', 'code_eoi', 'longitude',
       'latitude', 'location', 'district_code', 'district_name', 
       'neighborhood_code', 'neighborhood_name', 'class_1', 'class_2',  
       'contaminant_1', 'contaminant_2', 'contaminant_3']

df_2018.columns = cols_2018

df_2018
```

```python
# 7 stations and 16 columns
df_2018.shape
```

```python
stn.print_categorical_cols(df_2018, df_2018.columns)
```

## Stations in the year 2019

```python
df_2019
```

```python
df_2019.shape
```

```python
stn.print_categorical_cols(df_2019, eda_cols)
```

```python
stn.print_individual_col(df_2019, 'longitude')
stn.print_individual_col(df_2019, 'latitude')
```

```python
stn.contaminants_df
```

```python
d_cond_2019, df_cond_2019 = stn.condense_stations_data(df_2019, stn.contaminants_df)
```

```python
df_cond_2019
```

```python
data_2019 = {}
```

```python
for index, row in df_cond_2019.iterrows():
    cols = list(df_cond_2019.columns)[1:]
    data_2019[row['station']] = {str(x): row[x] for x in cols}
```

```python
#list(df_cond_2019.columns)[1:]
important_cols = [
    ('station_name', 'Station Name'),
    ('location', 'Location'),
    ('district_code', 'District Code'),
    ('district_name', 'District Name'),
    ('neighborhood_code', 'Neighborhood Code'),
    ('neighborhood_name', 'Neighborhood Name'),
    ('class_1', 'Station Class'),
    ('class_2', 'Station Sub-Class'),
    ('longitude', 'Longitude'),
    ('latitude', 'Latitude'),
    ('contaminant_code', 'Contaminants Codes'),
    ('contaminant_names', 'Contaminants Names')]
```

```python
map_2019 = folium.Map(
    location = [df_cond_2019.latitude.mean(), df_cond_2019.longitude.mean()], 
    tiles = "OpenStreetMap", 
    zoom_start = 12
)
```

```python
for k in data_2019.keys():
    msg = '<div style="min-width: 400px;" ><b>Station Id: ' + str(k) + '</b></br>'
    msg += '<ul>'
    for x in important_cols:
        msg += '<li><b><i>' + str(x[1]) + '</i></b> -- "' + str(data_2019[k][x[0]]) + '"'
    msg += '</ul></div>'

    #print(msg)
    
    folium.Marker(
      location = [data_2019[k]['latitude'], data_2019[k]['longitude']],
      #popup = [data_2019[k]['station_name'], data_2019[k]['latitude'], data_2019[k]['longitude']],
      popup = msg,
   ).add_to(map_2019)
```

```python
map_2019
```

```python
map_2019.save('.\interative_maps\map_2019.html')
```
